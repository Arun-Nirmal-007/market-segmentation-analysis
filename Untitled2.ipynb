# Step 9 :  Customising the Marketing Mix

```python
# Adjust the 'Like' score for frequent visitors
df['adjusted_like'] = df.apply(lambda row: row['Like'] if row['VisitFrequency'] == 'Once a week' else 0, axis=1)

# Display the updated DataFrame with the new 'adjusted_like' column
df[['VisitFrequency', 'Like', 'adjusted_like']].head()
```

```python
import matplotlib.pyplot as plt

# Assuming k4 is a cluster assignment variable you have calculated
visit = df.groupby('k4')['VisitFrequency'].mean()
like = df.groupby('k4')['Like.n'].mean()
female = df.groupby('k4').apply(lambda x: (x['Gender'] == "Female").mean())

# Plotting
plt.scatter(visit, like, s=10*female, xlim=(2, 4.5), ylim=(-3, 3))
for i in range(len(visit)):
    plt.text(visit[i], like[i], str(i+1))
plt.xlabel('Visit Frequency')
plt.ylabel('Like')
plt.show()
```
